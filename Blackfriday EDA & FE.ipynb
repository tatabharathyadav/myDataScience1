##Black fiday dataset EDA and feature engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train=pd.read_csv('blackfridaytrain.csv')
df_train.head()
```

```python
df_test=pd.read_csv('blackfridaytest.csv')
df_test.head()
```

```python
df=df_train.append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
del df['User_ID']  ##df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#categorical data
```

```python
pd.get_dummies(df['Gender'],drop_first=True).head()
```

```python
df['Gender']=df['Gender'].replace({'F':0,'M':1})    #we can use map also inplace of replace
df.head()
```

```python
df['Age'].unique()
```

```python
df['Age'].value_counts().plot(kind='bar')
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5, '51-55':6,'55+':7})
df.head()
```

```python
#handling city_category
```

```python
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
del df['City_Category']
```

```python
df.head()
```

```python
###Handling missing values
```

```python
df.isnull().sum()
```

# replace missing values with mode if it is categorical data or descrete data

```python
df['Product_Category_2'].unique()   ##it is descrete data
```

```python
df1=df['Product_Category_2'].mode()[0]
df1
```

```python
df['Product_Category_2']=df['Product_Category_2'].fillna(df1)
df['Product_Category_2']
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
df.head()
```

```python
df2=df['Product_Category_3'].mode()[0]
df2
```

```python
df['Product_Category_3']=df['Product_Category_3'].fillna(df2)
df['Product_Category_3']
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
df.head()
```

```python
df.shape
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
df.info()
```

```python
df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
df.info()
```
