```python
import yfinance as yf
import datetime

start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)

# 'BAC' is the ticker symbol for Bank of America
BAC = yf.download('BAC', start=start, end=end)
AAPL = yf.download('AAPL', start=start, end=end)
MSFT = yf.download('MSFT', start=start, end=end)
AMZN= yf.download('AMZN', start=start, end=end)
TSLA= yf.download('TSLA', start=start, end=end)
KO= yf.download('KO', start=start, end=end)
print(KO)
```

```python
del BAC['Adj Close']
```

```python
BAC
```

```python
tickers=['BAC','AAPL','MSFT','AMZN','TSLA','KO']
tickers
```

```python
bank_stocks=pd.concat([BAC,AAPL,MSFT,AMZN,TSLA,KO],axis=1,keys=tickers)
bank_stocks.head()
```

```python
bank_stocks.columns.names=['BankTicker','StockInFo']
bank_stocks
```

```python
for tick in tickers:
    print(tick,bank_stocks[tick]['Close'].max())
```

```python
bank_stocks.xs(key='Close',axis=1,level='StockInFo').max()
```

```python
returns=pd.DataFrame()
for tick in tickers:
    returns[tick +' Return']=bank_stocks[tick]['Close'].pct_change()
returns.head()
```

```python
import seaborn as sns
sns.pairplot(returns[1:])
```
